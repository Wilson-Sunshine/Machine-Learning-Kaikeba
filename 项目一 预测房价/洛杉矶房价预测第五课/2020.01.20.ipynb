## 相关性分析

相关性系数 r in [-1,+1]

|r| > 0.95 显著关系,相互之间的关系具有较强关联,甚至于简单的线性计算关系或者是同构关系,如果是重合度很高的数据,可以考虑去掉其中一个,或者进行融合

|r| >= 0.8 强相关

|r| >= 0.5 中等相关

|r| >= 0.3 弱相关,特征属于有待挖掘,可以通过特征工程对于特征进行优化或者是合并构造

|r| < 0.3 极弱相关,可以考虑去掉
