### Seaborn绘图工具与数据分析基础

基于pyplot

自带多类数据集,iris,Titanic,tips,mnist....

对于图形绘制进行了优化

具有高级数据统计接口


```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### 1.消费额与天(周一...周日)之间的关系--带状图

```python
sns.stripplot(data=tips,x='day',y='total_bill',jitter=True)
```

### 2. 小费与天之间的关系--蜂群图

```python
sns.swarmplot(data=tips,x='day',y='tip' )
```

### 3.消费、天以及就餐时间三者关心--Hue

```python
sns.swarmplot(data=tips,x='day',y='total_bill',hue='time')
```

该餐馆距离工作地比较近,同时距离消费和消遣的地方也比较近

### 4.买单者与性别的关系

```python
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex')
```

工作日的男女买单数量基本一致,AA制工作餐,或者是各吃各的;周六日的时候男士买单比较多,因为关系比较密切

```python
sns.swarmplot(data=tips,x='day',y='total_bill',hue='size')
```

```python
sns.swarmplot(data=tips,x='size',y='total_bill')
```

```python
sns.swarmplot(data=tips,x='day',y='tip',hue='size')
```

### 每天的消费额度三分位数,最大值,中位数,一分位数,最小值--箱线图

```python
sns.boxplot('day','total_bill',data=tips)
```

```python
sns.boxplot('day','total_bill',data=tips,hue='time')
```

消遣的时候往往消费的Range会比较大

工作时候相对Range会比较小

```python
sns.boxplot('day','tip',data=tips,hue='time')
```

周末小费普遍偏高,且会有惊喜

### 匹萨饼的Size与消费额在不同消费时间的核密度函数

```python
sns.violinplot('size','total_bill',data=tips,hue='time')
```

### 是否抽烟与买单额度以及性别的关系--swarm-violin图

```python
sns.violinplot('smoker','total_bill',data=tips)

sns.swarmplot('smoker','total_bill',data=tips,color='w')
```

### 核密度估计KDE

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

total_bill本身属于左偏分布,如果想要该特征变得更加规范化,可以通过log或者sqrt进行降阶,可以适当解决偏度问题

```python
sns.distplot(np.log(tips['total_bill']))
```

```python
sns.distplot(np.sqrt(tips['total_bill']))
```

不一定取log做成正太分布就是最好的,因为取log之后可能会丢掉相关性的信息

### tip与size的关系---带有线性回归的散点图--lmplot

```python
sns.lmplot('size','tip',data=tips)
```

```python
sns.lmplot('tip','total_bill',data=tips)
```

```python
tips.corr() #相关性系数
```

```python
tips['tip'].corr(tips['total_bill'])
```

### total_bill与tip之间的相关性以及各自变量的核密度分布---jointplot

```python
sns.jointplot('total_bill','tip',data=tips,kind='reg')
```

根据此图可以得到total_bill以及tip的相关性核密度分布
