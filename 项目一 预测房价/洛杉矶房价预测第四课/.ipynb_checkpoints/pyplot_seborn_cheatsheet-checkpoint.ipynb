### 结构化数据绘制散点图

```python
data = {
    'a':np.arange(50),
    'c':np.random.randint(0,50,50),
    'd':np.random.randn(50),
}
```

```python
data
```

### plt.scatter()散点图

```python
plt.scatter('a','c',data=data)
plt.xlabel('a data')
plt.ylabel('c data')
```

```python
plt.scatter('a','c',data=data,c='c',marker='>')
plt.xlabel('a data')
plt.ylabel('c data')
```

### 柱状图

```python
names=['A','B','C']

values=[1,100,1000]

plt.bar(names,values)
```

```python
plt.bar(range(len(names)),values)
plt.xticks(range(len(names)),names)
```

### 子图绘制-subplot

1.讲画布进行切分

2.将plot图分配到指定的位置和大小


```python
plt.figure(1,figsize=(9,4))
plt.subplot(131)
plt.bar(names,values,color='r')

plt.subplot(132)
plt.bar(names,values,color='g')

plt.subplot(233)
plt.bar(names,values,color='g')
```

### Seaborn绘图工具与数据分析基础

基于pyplot

自带多类数据集,iris,Titanic,tips,mnist....

对于图形绘制进行了优化

具有高级数据统计接口


```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### 1.消费额与天(周一...周日)之间的关系--带状图

```python
sns.stripplot(data=tips,x='day',y='total_bill',jitter=True)
```

### 2. 小费与天之间的关系--蜂群图

```python
sns.swarmplot(data=tips,x='day',y='tip' )
```

### 3.消费、天以及就餐时间三者关心--Hue

```python
sns.swarmplot(data=tips,x='day',y='total_bill',hue='time')
```

该餐馆距离工作地比较近,同时距离消费和消遣的地方也比较近

### 4.买单者与性别的关系

```python
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex')
```

工作日的男女买单数量基本一致,AA制工作餐,或者是各吃各的;周六日的时候男士买单比较多,因为关系比较密切

```python
sns.swarmplot(data=tips,x='day',y='total_bill',hue='size')
```

```python
sns.swarmplot(data=tips,x='size',y='total_bill')
```

```python
sns.swarmplot(data=tips,x='day',y='tip',hue='size')
```

### 每天的消费额度三分位数,最大值,中位数,一分位数,最小值--箱线图

```python
sns.boxplot('day','total_bill',data=tips)
```
