根据如上的缺失率分析可以选择不同的填充方案(粗填充)

.1  缺失率大于80%的列,可以选择去掉,原因是缺失值可能引入更大的噪声-(可能缺失是合理的,但是去掉之后,其实去掉了更多的有效数据)

.2  使用平均数/中位数/众数(离散型数据)进行填充

粗填充绝对不是最终的填充方案,但是从算法迭代的角度,需要快速形成一个可被学习的训练集,所以优先选择快速填充

粗填充是为了跑第一个baseline算法而使用的

### 不同类型的填充

Numerical|categorial

Numerical

1.均值填充

2.中位数填充

4.内在关联填充-EDA的范畴-最有效的填充

Categorial

```python
train1.mean()
```

```python
train1.median()
```

### 对于数值型数据进行均值| 中位数填充

```python
train1 = train1.fillna(train1.median()) # 用中位数填充
```

```python
tmp = train1.isnull().sum().sort_values()
```

```python
tmp[tmp>0].shape
```

### 对于离散型数据进行填充

朴素办法: NaN数据使用“None”数据类型来进行填充或者“NA”

```python
train1 = train1.fillna('None')
```

```python
tmp = train1.isnull().sum().sort_values()
```

```python
tmp[tmp>0].shape
```

### BaseLine模型的训练


```python
import xgboost as xgb
```

```python
xg_reg = xgb.XGBRegressor(objective = 'reg:linear',colsample_bytree=0.7,learning_rate = 0.01,max_depth=5,n_estimators=3700,subsample=0.8,random_state=123)
```

### 训练集的准备

TrainSet ValidSet TestSet

TrainSet 通常是有Label的数据集

ValidSet 通常也是有Label的,但是不参与训练,但是会在训练中对于训练过程的模型进行评测,来衡量和记录Test-Error和Test Loss 的变化(valid-acc/valid-loss)

TestSet 有可能没有Label


### 将Pandas中的Object数据进行数值化,且离散的数据进行0/1二值化

```python
train1.head()
```

```python
train1.shape
```

如上所述,首先Object类型的数据,应该被数值(离散型的数值)所代替,因为对于计算型的模型,需要用到数值特征

其次,离散型用来描述类型的数值本身是没有大小关系的,但是这些数值引入到计算型的模型中(线性类,神经网络类,SVM类)就会引入噪声

因此还要对于离散型的数据进行0/1数值化


```python
X = pd.get_dummies(train1)
```

```python
X.shape
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import mean_squared_error
```

```python
### x中不能出现Target,将Target变量分离y

y=X['SalePrice']

X=X.drop(['Id','SalePrice'],axis=1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 123)
```

```python
X_train.shape
```

```python
X_test.shape
```

### 使用XGBRegressor进行回归

```python
xg_reg.fit(X_train,y_train)
```

### 使用Test数据集进行测试

```python
pred = xg_reg.predict(X_test)
```

```python
pred[:5]
```

```python
rmse = np.sqrt(mean_squared_error(y_test,pred)) #标准差
```

```python
rmse
```

```python
logrmse = np.sqrt(mean_squared_error(np.log(y_test),np.log(pred)))
```

```python
logrmse
```

## 数据 EDA的入门

1.数据属性分布分析(属性本身的概率密度分析/正态分布分析),当数据是偏态过于严重的时候,可能会造成计算时精度的损失,必要的时候可以进行标准化的操作

2.相关性分析(特征列跟目标之间的关系) - 相关性系数 - correlationship

3.缺失属性的合理化填充

4.构造属性和特征工程



## 相关性系数

相关性系数 r in [-1,+1]

|r| > 0.95 显著关系,相互之间的关系具有较强关联,甚至于简单的线性计算关系或者是同构关系,如果是重合度很高的数据,可以考虑去掉其中一个,或者进行融合

|r| > 0.8 强相关

|r| = 0.5 中等相关

|r| = 0.3 弱相关,特征属于有待挖掘,可以通过特征工程对于特征进行优化或者是合并构造

|r| < 0.3 极弱相关,可以考虑去掉
